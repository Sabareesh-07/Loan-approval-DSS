# Reading Dataset1.csv
This notebook reads and displays the first 10 rows from dataset1.csv

```python
# Import required library
import pandas as pd
```

```python
# Read and display the first 10 rows of the dataset
try:
    # Read the CSV file
    df = pd.read_csv('dataset1.csv')
    
    # Display the first 10 rows
    print("First 10 rows of dataset1.csv:")
    print(df.head(10))
    
except FileNotFoundError:
    print("Error: dataset1.csv file not found in the current directory")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
